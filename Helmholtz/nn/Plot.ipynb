```python
import sys
import numpy as np
sys.path.append('../../../nn')

N_theta = 100

prefix = "./"
theta_field = np.load(prefix+"Random_Helmholtz_theta_" + str(N_theta) + ".npy")   
K_field = np.load(prefix+"Random_Helmholtz_K_" + str(N_theta) + ".npy")
c_field = np.load(prefix+"Random_Helmholtz_cs_" + str(N_theta) + ".npy")   
```

```python
import matplotlib.pyplot as plt
######################################################
N_data, N_theta  =  theta_field.shape
N_x, N_y, N_data = K_field.shape

L = 1.0
xx = np.linspace(0, L, N_x)
Y, X = np.meshgrid(xx, xx)


i = 1

vmin, vmax = None, None
fig = plt.figure()
plt.pcolormesh(X, Y, K_field[:, :, i], shading="gouraud", vmin=vmin, vmax=vmax)
plt.colorbar()
plt.title("K")

fig = plt.figure()
plt.pcolormesh(X, Y, c_field[:, :, i], shading="gouraud", vmin=vmin, vmax=vmax)
plt.colorbar()
plt.title("cs")
```

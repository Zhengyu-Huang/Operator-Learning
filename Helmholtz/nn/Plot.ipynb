```python
import sys
import numpy as np
sys.path.append('../../../nn')

N_theta = 100

prefix = "./"
theta_field = np.load(prefix+"Random_Helmholtz_theta_" + str(N_theta) + ".npy")   
K_field = np.load(prefix+"Random_Helmholtz_K_" + str(N_theta) + ".npy")
c_field = np.load(prefix+"Random_Helmholtz_cs_" + str(N_theta) + ".npy")   
```

```python
import matplotlib.pyplot as plt
######################################################
N_data, N_theta  =  theta_field.shape
N_x, N_y, N_data = K_field.shape

L = 1.0
xx = np.linspace(0, L, N_x)
Y, X = np.meshgrid(xx, xx)


for ind in [0, 10]:
    
    fig,ax = plt.subplots(ncols=2, figsize=(12,5))
    im0 = ax[0].pcolormesh(X, Y, c_field[:, :, ind],  shading="gouraud")
    fig.colorbar(im0, ax=ax[0])
    im1 = ax[1].pcolormesh(X, Y, K_field[:, :, ind],   shading="gouraud")
    fig.colorbar(im1, ax=ax[1])
    
    plt.tight_layout()
    
    plt.savefig('Helmholtz-map%d.png' %(ind))

```

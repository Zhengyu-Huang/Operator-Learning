```python
import sys
import numpy as np
sys.path.append('../../../nn')

N_theta = 100

prefix = "./"
curl_f_field = np.load(prefix + "Random_NS_curl_f_" + str(N_theta) + ".npy")   
omega_field = np.load(prefix + "Random_NS_omega_"  + str(N_theta) + ".npy")
theta_field = np.load(prefix+ "Random_NS_theta_" + str(N_theta) + ".npy")   
```

```python
import matplotlib.pyplot as plt
######################################################
N_data, N_theta  =  theta_field.shape
N_x, N_y, N_data = omega_field.shape

L = 2*np.pi
xx = np.linspace(0, L, N_x)
Y, X = np.meshgrid(xx, xx)


for i in range(5):

    vmin, vmax = None, None
    fig = plt.figure()
    plt.pcolormesh(X, Y, omega_field[:, :, i], shading="gouraud", vmin=vmin, vmax=vmax)
    plt.colorbar()
    plt.title("omega")

    fig = plt.figure()
    plt.pcolormesh(X, Y, curl_f_field[:, :, i], shading="gouraud", vmin=vmin, vmax=vmax)
    plt.colorbar()
    plt.title("curl_f")
    
```
